```
import pandas as pd
```

```
# File location and type
file_location = "/FileStore/tables/trip_data.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
tripData_df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(tripData_df)
```

```
# File location and type
file_location = "/FileStore/tables/station_data.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
StationData_df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(StationData_df)
```

```
# 2. Cuantos viajes son menores a media hora?
```

```
# Convertir la columna Duration al tipo de datos entero
tripData['Duration'] = tripData['Duration'].astype(int)
# Implementar tu codigo aqui
viajes_menos_30min = tripData[tripData['Duration'] < 1800] # 30 min es 1800 seg.
num_viajes_menos_30min = len(viajes_menos_30min)

print("El nmero de viajes menores a 30 minutos es:", num_viajes_menos_30min)
```

```
# 3. Cuantos viajes comenzarn en la estacion inicial "Post at Kearney" ?
##### Unir los datasets
```

```
tripData['Start Terminal'] = tripData['Start Terminal'].astype(float)
StationData['station_id'] = StationData['station_id'].astype(float)
```

```
data_merged_start_terminal = pd.merge(StationData, tripData, left_on='station_id', right_on='Start Terminal')
data_merged_start_terminal.head()
```

```
# Implementar tu codigo aqui
conteo_viajes_estacion_inicial = data_merged_start_terminal[data_merged_start_terminal['name'] == "Post at Kearney"]
conteo_viajes_estacion_inicial.count()
```

```
# 4. Cuantos viajes terminaron en la ciudad de "San Francisco"?
```

```
tripData['End Terminal'] = tripData['End Terminal'].astype(float)
data_merged_end_terminal = pd.merge(StationData, tripData, left_on='station_id', right_on='End Terminal')
data_merged_end_terminal.head()
```

```
# 7. En que dia hubiern ms viajes ?
```

```
#COPIA DEL DATAFRAME tripData
tripData_cy = tripData.copy()
tripData_cy
```

```
tripData_cy['Date'] = pd.to_datetime(tripData_cy['Start Date'])
tripData_cy['Dia'] = tripData_cy['Date'].dt.day_name()
tripData_cy
```

```
# Implementar tu codigo aqui
grupo_Dia = tripData_cy.groupby('Dia').count().idxmax()
grupo_Dia
```

```
# 8. Cual es el dia donde el promedio del viaje es mayor?
```

```
# Implementar tu codigo aqui
grupo_Dia_average = tripData_cy.groupby('Dia').mean().idxmax()
grupo_Dia_average
```
